```python
## json loader and csv file writer

import json
import csv

# Open the result.json file and load its contents into a dictionary
with open('result.json', 'r') as f:
    json_data = json.load(f)

# Open the listing.csv file and create a CSV writer
with open('listing.csv', 'w', newline='') as f:
    writer = csv.writer(f)

    # Write the header row to the CSV file
    writer.writerow(['id', 'rating'])

    # Loop through each row in the JSON data
    for row in json_data:
        # Get the ID and rating from the JSON data
        id = row['id']
        rating = row['label']

        # Find the corresponding row in the listing.csv file
        with open('listing.csv', 'r') as csvfile:
            reader = csv.reader(csvfile)
            next(reader)  # skip the header row
            for csv_row in reader:
                if csv_row[0] == id:
                    # Write the ID and rating to the CSV file
                    writer.writerow([id, rating])
                    break
```

```python
#Working Version!! with smaller file size##

import csv
import os
import requests
from multiprocessing.pool import ThreadPool
from PIL import Image
from io import BytesIO

# Set directory to save images
SAVE_DIR = "low_images"

# Create save directory if it doesn't exist
if not os.path.exists(SAVE_DIR):
    os.makedirs(SAVE_DIR)

# Define a function to download and resize an image
def download_image(row):
    url = row["picture_url"]
    filename = str(row.get("id", "")) + ".jpg"
    # Download image from URL
    response = requests.get(url)
    # Check that response contains image data
    content_type = response.headers.get('Content-Type')
    if 'image' not in content_type:
        print(f"{url} is not an image")
        return
    # Open image with Pillow
    try:
        img = Image.open(BytesIO(response.content))
    except:
        print(f"{url} is not a valid image file")
        return
     # Convert image to RGB mode if it is in RGBA mode
    if img.mode == 'RGBA':
        img = img.convert('RGB')
    # Resize image to 50% of its original size
    width, height = img.size
    new_size = (int(width * 0.5), int(height * 0.5))
    img = img.resize(new_size, resample=Image.LANCZOS)
    # Save image to file
    with open(os.path.join(SAVE_DIR, filename), 'wb') as f:
        img.save(f, "JPEG")

# Load CSV file
with open('/Users/euisuck2/Downloads/listings-2.csv', 'r',encoding='utf-8-sig') as csv_file:
    reader = csv.DictReader(csv_file)
    # Download images using ThreadPool
    pool = ThreadPool()
    pool.map(download_image, reader)
    pool.close()
    pool.join()

```
